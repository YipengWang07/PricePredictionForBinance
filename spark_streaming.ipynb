```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
from pyspark import SparkConf
from pyspark import SparkContext, HiveContext
from pyspark.streaming import StreamingContext
from pyspark.sql import SQLContext
from pyspark.sql.functions import split, unix_timestamp, from_unixtime, to_timestamp, window, col
from pyspark.sql import functions as F
import pyspark

from pyspark.sql.types import StringType, IntegerType, StructType, StructField, FloatType

import json
```

```python
sc = SparkContext("local[*]", "twitter_streaming")

sqlContext = SQLContext(sc)
spark = sqlContext.sparkSession
```

```python
lines = spark \
    .readStream \
    .format("socket") \
    .option("host", "localhost") \
    .option("port", 5544) \
    .load()
```

```python
split_col = split(lines.value, ',',)

df_in = lines \
    .withColumn("timestamp", split_col.getItem(0)) \
    .withColumn("flw_count", split_col.getItem(1).cast(IntegerType())) \
    .withColumn("score", split_col.getItem(2).cast(FloatType())) \
    .drop("value")

df = df_in \
    .withColumn("multiplied_score", df_in.score * df_in.flw_count) \
    .withColumn("created_time", to_timestamp(from_unixtime(unix_timestamp('timestamp', 'EEE MMM d HH:mm:ss z yyyy')))) \
    .withWatermark("created_time", "1 minutes")

group_df = df.groupBy(window("created_time", "1 minute")) \
    .agg(F.mean("score").alias("score"), 
         F.sum("multiplied_score").alias("multiplied_score"), 
         F.sum("flw_count").alias("flw_count")) \
    .withColumn("weighted_score", col("multiplied_score") / col("flw_count")) \
    .withColumn("start_time", col("window.start")) \
    .withColumn("end_time", col("window.end")) \
    .drop("window", "multiplied_score", "flw_count")
```

```python
query = group_df \
    .writeStream \
    .format("parquet") \
    .outputMode("append") \
    .option("path", "./twitter_score") \
    .option("header", True) \
    .option("checkpointLocation", "./checkpoint") \
    .start() \
    .awaitTermination()
```
